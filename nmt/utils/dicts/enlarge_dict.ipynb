```python
import os,sys,pickle
from tqdm import tqdm
```

```python
old_comp_dict=pickle.load(open('./comp_dict_1.pkl','rb'))
old_stroke_dict=pickle.load(open('./stroke_dict_2.pkl','rb'))
```

```python
char_comp_dict={old_comp_dict[s]:s for s in old_comp_dict}
char_stroke_dict={old_stroke_dict[s]:s for s in old_stroke_dict}
```

```python
mecab_vocab_set=set(w for l in open('/work/vincentzlt/tf-nmt/data_2/mecab-jieba/train.jp','rt') for w in l.split())
jieba_vocab_set=set(w for l in open('/work/vincentzlt/tf-nmt/data_2/mecab-jieba/train.cn','rt') for w in l.split())
```

```python
comp_dict=old_comp_dict.copy()
stroke_dict=old_stroke_dict.copy()
```

```python
old_vocab_set=set(comp_dict[s] for s in comp_dict)
```

```python
for w in tqdm(mecab_vocab_set):
  if not w in old_vocab_set:
    s_stroke=''.join(char_stroke_dict[c] for c in w)
    s_comp=''.join(char_comp_dict[c] for c in w)
    if not s_comp in comp_dict:
      comp_dict[s_comp]=w
    else:
      idx=1
      while True:
        if not '{}_{}'.format(s_comp,idx) in comp_dict:
          comp_dict['{}_{}'.format(s_comp,idx)]=w
          break
        else:
          idx+=1
    if not s_stroke in stroke_dict:
      stroke_dict[s_stroke]=w
    else:
      idx=1
      while True:
        if not '{}_{}'.format(s_stroke,idx) in stroke_dict:
          stroke_dict['{}_{}'.format(s_stroke,idx)]=w
          break
        else:
          idx+=1
    
```

```python
old_vocab_set=set(comp_dict[s] for s in comp_dict)
```

```python
for w in tqdm(jieba_vocab_set):
  if not w in old_vocab_set:
    s_stroke=''.join(char_stroke_dict[c] for c in w)
    s_comp=''.join(char_comp_dict[c] for c in w)
    if not s_comp in comp_dict:
      comp_dict[s_comp]=w
    else:
      idx=1
      while True:
        if not '{}_{}'.format(s_comp,idx) in comp_dict:
          comp_dict['{}_{}'.format(s_comp,idx)]=w
          break
        else:
          idx+=1
    if not s_stroke in stroke_dict:
      stroke_dict[s_stroke]=w
    else:
      idx=1
      while True:
        if not '{}_{}'.format(s_stroke,idx) in stroke_dict:
          stroke_dict['{}_{}'.format(s_stroke,idx)]=w
          break
        else:
          idx+=1
```
